<h3>
Sampled assessments scenario
</h3><br>
Then I try to sample out some random assessments from the previous scenario, to simulate the fact that not all students may have undergone assessment for each course in which they are enrolled. I decide a percentage of enrollments that I want to sample out from the <i>grading</i> table. Then I compute its inverse, which is in fact the percentage of enrollments that I want to have undergone assessment and I translate that percentage into an integer number, that is the exact number of elements that I want to keep in the <i>grading</i> table. However, it works better to create a new dictionary and append to its lists these elements. Then I will use this dictionary instead of the original one.

```python
# I want 20% of all of the enrollments not to have an assessment (which means I want 80% to have an assessment)
end = len(gradingDict['student'])
sampleIn = list()
while len(sampleIn) < int(end * 0.8):
    randomNumber = randint(0, end)
    if randomNumber not in sampleIn:
        sampleIn.append(randomNumber) 
# notice that sampling is without replacement, hence there can be int(end * 0.8) gradings, one per enrollment
```

I generate a new empty dictionary with the same keys as the gradingDict and fill it with elements from gradingDict located at the positions stored in the list <i>sampleIn</i>. This will be my new grading table.

```python
newGradingDict = dict()
for k in gradingDict.keys():
    newGradingDict[k] = list()
```

```python
for i in range(0, end):
    if i in sampleIn:
        newGradingDict['course'].append(gradingDict['course'][i])
        newGradingDict['student'].append(gradingDict['student'][i])
        newGradingDict['assessment'].append(gradingDict['assessment'][i])
        newGradingDict['score'].append(gradingDict['score'][i])
        newGradingDict['attemptDate'].append(gradingDict['attemptDate'][i])
```

<h3>
Multiple sampled assessments scenario
</h3><br>
Now I want to implement the possibility that a student has undergone assessment more than once for a course to which she is enrolled, because its grading score was not sufficient to obtain the course certificate. I consider only tuples with grading scores less than 60 and sample some of them to replicate studentID, courseID and assessmentID but with a new score (not necessarily different from the one already in newGradingDict) and a new date, necessarily following the one already present in newGradingDict and preceding august 31st, 2023.

First of all I count the number of grading tuples with scores under 60. Then I compute a maximum percentage of tuples with scores under 60 where the student may have been assessed more than once, say 80%. Then I generate random numbers up to that 80%. Then I perform iterations in the new grading table and duplicate grading tuples if the ith iteration is equal to one of those random numbers.

```python
counter2 = 0
for i in newGradingDict['student']:
    counter2 += 1
print(counter2)
```

Then I compute the 80% of the counter (as an integer) and fill a list with random numbers up to that integer. Random numbers must not be replicated within the list, so that I obtain exactly a list length of 80% the value of the counter.

```python
sampleDuplicate = list()
while len(sampleDuplicate) < int(counter2 * 0.8):
    randomDuplicate = randint(0, counter2)
    if randomDuplicate not in sampleDuplicate:
        sampleDuplicate.append(randomDuplicate)
print(len(sampleDuplicate))
print(sampleDuplicate, end = '')
```

Then I iterate through the grading table until a length of 80% of the counter would be reached. When the value of 'score' within the actual grading tuple is less than the threshold and the position is included in the 80% sample list, I generate a new score and a date later than the previous assessment. I introduce October 15th, 2023, otherwise there is the chance that the previous assessment had taken place in August and would exceed the previously fixed limit date of August 31st, 2023.
It is to be noticed that, since two conditions are needed (the score to be under 60 and the position to be included in the 80% sample), it is hard that all 80% tuples sampled would respect both conditions. Hence, the number of extra assessments should normally be well under 80% of the insufficient scores and there's no way to predict this number exactly.

```python
counter3 = 0
for c, s, a, sc, ad in zip(newGradingDict['course'], newGradingDict['student'], newGradingDict['assessment'], newGradingDict['score'], newGradingDict['attemptDate']):
    if sc < 60 and counter3 in sampleDuplicate:
        newGradingDict['course'].append(c)
        newGradingDict['student'].append(s)
        newGradingDict['assessment'].append(a)
        newGradingDict['score'].append(randint(40, 100))
        strDate = str(ad)
        dateList1 = strDate.split('-')
        year1 = int(dateList1[0])
        month1 = int(dateList1[1])
        day1 = int(dateList1[2])
        newGradingDict['attemptDate'].append(fake.date_between(datetime.date(year1, month1, day1) + datetime.timedelta(15), datetime.date(2023, 10, 16)))
    counter3 += 1
```
