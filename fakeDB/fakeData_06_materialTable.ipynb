<h4>
Create course units
</h4><br>
A first step in creating the learning material is to identify the course unit to which they belong. The most straightforward approach to perform this identification is to first define the units for each course. This can be done by randomly attaching a random number of units per course and generating them through a loop over course. A unit, within a course, is simply a string attaching a progressive integer to the word 'Unit '. I don't want courses to be too long, so I plan them to be around 4-5 units long on average. But I also imagine the existence of some outlier courses up to 8 units. In order to be allowed to randomly choose the course length in units with a prevalence of 4-5 units courses I create a list of 10 integers from 4 to 8 where the 70% of occurrences is a 4 or a 5 and the remaining 30% is a 6, a 7 or a 8. Then, for each course in my course table, I sample one of the integers in the list and create as many course units as the integer sampled from the list.

<h4>
Assign learning material to units
</h4><br>
Now that we have course units, we can create the learning material for each course unit. The learning material is composed by video lectures and relative slides and tests. There are usually multiple video lectures and slides, and these appear with the same frequency, since slides reproduce the on-screen content of the video lectures. Tests are usually submitted at the end of each unit and consist of multiple-choice quizzes. <br> A huge challenge, here, is to give a name to the learning material. While for unit tests this can be pretty straightforward (we might simply attach the word 'test' to the unit name, or compose it by concatenating the word 'End of ' to the lowecase unit name and then to the word ' test'), it can be hard for video lectures and slides (the challenge is not double, anyway, since if we find a suitable name for a video lecture, this can be replicated for its slides, or viceversa). To make it easier, we could extrapolate a couple of keywords from the course name and use them in conjunction with the words 'introduction', or 'advanced subjects', etc. But it would be impossible to obtain a decent variability in the material name, if we think that a course subject may need to learn basics of other subjects thet can never be cited in the course title. Or that the process of learning a course subject may undergo subject specifications that are too subtle to be part of the course title. <br> A way to address this challenge may exploit the discipline to which a course belongs. So that the first units may deal with generic subjects in the discipline and the later units may refer to the specific course subjet. It doesn't look easy anyway. <br> A radical way to address the challenge would be to generate completely random sentences for each learning material. This way could never guarantee that the material name would have anything to do with the course subject, if it even made any sense. <br> A final approach would be to store material names into a text file, categorize them by discipline and randomly attach them to a course material based on the discipline. This would help in making the learning-material name positively relate to the course subject, even if there would be various inconsistencies. <br> But we are building this database just to perform queries and test query execution times, not for any practical purpose. So the last two approaches seem perfectly legitimate.

Now each unit in the course must be filled up with learning materials. Obviously, each unit belongs to a course which is associated to a specific discipline. The same association exists for the learning material, hence a unit must be filled up with learning material where the associated discipline is the same as that of the course to which the unit belongs. This is the first requisite that the learning material must have. Units have neutral names and, moreover, they are not independent constructs, they are simply attributes in the learning material table. So the only way to correctly associate a learning material to a unit is through the course ID. Each course has a potentially different number of units, so, although unit names are interchangeable across courses, I need to draw unit names from the correct course because otherwise I risk to associate a learning material of <i>economics and management</i> to a unit number that does not exists in courses of <i>economics and management</i>. The best way not to lose track of the creation of the learning material is to start from a course (the course ID, actually) and then move to its units. When looping across the units of a course I can keep the course ID, the unit name and associate a learning material ID, type and name (and dimension) to the first two attributes.<br>A choice to be made at this point is to decide if each unit must have the same number of learning material or if to create a random number (with a minimum and a maximum, as for what has been done for course units) of learning material for each unit. Since the second option seems more realistic and it is mostly effortless in terms of code to produce, I opt to create a random number of <i>lm</i> items for each course unit. Of course, the same learning material must not repeat itself within the same course unit (and, possibly, neither across units of the same course).<br>As long as learning materials are concerned, I find it reasonable to think that the variability in the quantity of learning material per unit is higher than that of units per course. Hence I build another list of numbers in the range 4 - 10, from which to pick the number of learning material for the unit I'm looping through. I want the average number of learning material to lean on the left side of the sequence, hence smaller numbers (4, 5) must appear more frequently than lager ones (7, 8, 9, 10). Consider, though, that I create two learning materials for each unit in the randomly chosen number and I add a final unit quiz. Hence, for each number <i><code>p</code></i> in the list from 4 to 10, I create <code>(p * 2) + 1</code> learning materials.

<h3>
Generate fake material data
</h3><br>
While generating the material table, I keep a list of course IDs that matches units' and learning materials' positions, so that I can use it when building the learning table.

```python
randomListLM = [4, 10, 6, 8, 5, 5, 7, 4, 9, 5]
longCourseList = list()
materialID = 0
for course, courseDisc, unit in zip(unitDict['course'], unitDict['courseDiscipline'], unitDict['unit']):
    i = 0
    randomNumber = sample(randomListLM, 1)[0]
    while i < randomNumber:
        lmIndex = sample(list(range(0, len(lmDict['lm']))), 1)[0]
        if courseDisc == lmDict['discipline'][lmIndex]:
            materialID += 1
            longCourseList.append(course)
            materialDict['materialID'].append(materialID)
            materialDict['unit'].append(unit)
            materialDict['materialType'].append('video lecture')
            materialDict['name'].append('[VIDEO] ' + lmDict['lm'][lmIndex])
            materialDict['dimension'].append(sample(list(range(0, 12)), 1)[0])
            materialID += 1
            longCourseList.append(course)
            materialDict['materialID'].append(materialID)
            materialDict['unit'].append(unit)
            materialDict['materialType'].append('lecture slides')
            materialDict['name'].append('[SLIDES] ' + lmDict['lm'][lmIndex])
            materialDict['dimension'].append(sample(list(range(0, 12)), 1)[0])
            i += 1
        else:
            i += 0
    materialID += 1
    longCourseList.append(course)
    materialDict['materialID'].append(materialID)
    materialDict['unit'].append(unit)
    materialDict['materialType'].append('quiz')
    materialDict['name'].append(unit + ' quiz')
    materialDict['dimension'].append(sample(list(range(0, 12)), 1)[0])
```
