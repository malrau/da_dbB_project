<h3>Routine to read an rtf file where there is a course name in each line </h3>

Define:
- legitimate characters in the file lines (only lines that start with these characters can be thought to store course names)
- path to the file to be read
- list where courses names will be stored

```python
# select file with courses and clean up lines where courses title are stored
legit = 'abcdefghijklmnopqrstuvwxyz '
path = '/Users/mau/OneDrive - unime.it/Learning/CdL Informatica/Anno II - Database/Module B/project/sources/[RAW]courses_and_lm/'
courseList = list()
```

<h4>Read the entire file</h4>

```python
with open(path + 'courses.rtf') as courses:
    print(courses.read())
```

<h4>Read the file line by line</h4>

```python
with open(path + 'courses.rtf') as courses:
    for line in courses:
        print(line)
```

<h4>Read the first three characters for each line</h4>
<h5>(so as to show which lines should be accepted in the final output)</h5>

```python
with open(path + 'courses.rtf') as courses:
    for line in courses:
        print(line[0 : 3])
```

<h4>Define a check for each line: if check is passed the line goes to the following process:</h4>

- first it is split in two based on the string 'from' (usually, after 'from' there is the university name)
- then only the first part of the splitted line is shown and appended to the course list

```python
with open(path + 'courses.rtf') as courses:
    for line in courses:
        check = 0
        for char in line[0 : 3]:
            if char in legit or char in legit.upper():
                check += 1
            else:
                check += 0
        if check == 3:
            line = line.split('from')
            print(line[0][0 : len(line[0]) - 1])
            courseList.append(line[0][0 : len(line[0]) - 1]) # the last character in the course name is a space
```

<h4>Check how many courses are stored in courseList, then remove those with <i>dirty</i> names</h4>

```python
print(len(courseList))
for course in courseList:
    if '\'' in course or '\'f' in course or '\'e' in course:
        courseList.remove(course)
print(len(courseList))
```

```python
for course in courseList:
    print(course)
```

<h4>Create course categories to be attached to each course</h4>

```python
# I decide the category to assign to a course based on keywords present in the course name
keyWordsDict = dict()

# list of keywords for the IT category
itKeyWords = ['programming', 'python', 'HTML', 'HTML5', 'Javascript', 'AI', 'TensorFlow ', 'Machine Learning', 'Deep Learning', 'program', 'cybersecurity', 'Data Science', 'GIS', 'SQL', 'database', 'algorithm', 'algorithms', 'algorithms,', 'algorithms:', 'algorithmic', 'computer', 'Google', 'Hadoop', 'Scala', 'MATLAB', 'web', 'developer', 'development', 'Big Data', 'internet', 'Deep Learning', 'Neural Networks', 'app', 'program', 'Java', 'cloud', 'visualisation', 'Tableau', 'design', 'software', 'cryptography', 'website', 'AutoCAD', 'Jira', 'AWS']
itKeyWordsUpper = list()
for word in itKeyWords:
    itKeyWordsUpper.append(word[0].upper() + word[1 : ])

# list of keywords for the maths category
mathKeyWords = ['mathematics', 'mathematical', 'logic', 'math', 'maths', 'linear algebra', 'calculus', 'model']
mathKeyWordsUpper = list()
for word in mathKeyWords:
    mathKeyWordsUpper.append(word[0].upper() + word[1 : ])

# list of keywords for the languages category
langKeyWords = ['english', 'chinese', 'korean', 'spanish', 'grammar', 'speaking', 'aprender', 'writing', 'write', 'language', 'linguistics', 'vocabulary', 'escritura', 'communication', 'communicating', 'tenses']
langKeyWordsUpper = list()
for word in langKeyWords:
    langKeyWordsUpper.append(word[0].upper() + word[1 : ])

# list of keywords for the psychology category
psychKeyWords = ['psychology', 'psychological', 'mindshift', 'mental', 'brain', 'success', 'success:', 'creative', 'creativos', 'parenting', 'emotion', 'emotions', 'influencing', 'human', 'meaning in life', 'potential', 'purpose', 'happiness']
psychKeyWordsUpper = list()
for word in psychKeyWords:
    psychKeyWordsUpper.append(word[0].upper() + word[1 : ])

# list of keywords for the economics and management category
econKeyWords = ['marketing', 'branding', 'brand', 'business', 'Game Theory', 'economics', 'manage', 'management', 'consumer', 'customer', 'capital', 'strategy', 'strategies', 'supply chain', 'Supply Chain', 'decision', 'entrepreneurship', 'leader', 'leading', 'finance', 'financial', 'finanzas', 'negocios', 'contabilidad', 'cryptocurrency', 'blockchain', 'negotiation:']
econKeyWordsUpper = list()
for word in econKeyWords:
    econKeyWordsUpper.append(word[0].upper() + word[1 : ])
    
# list of keywords for the sciences category
scienceKeyWords = ['research', 'physics', 'engineering', 'electronics', 'genetics', 'neurobiology', 'energy', 'chemistry', 'animal', 'dog', 'biology', 'genetics', 'Einstein:']
scienceKeyWordsUpper = list()
for word in scienceKeyWords:
    scienceKeyWordsUpper.append(word[0].upper() + word[1 : ])
    
# list of keywords for the medicine category
medicineKeyWords = ['epidemiology', 'cancer', 'clinical', 'health', 'weight', 'medical', 'medicine', 'COVID-19']
medicineKeyWordsUpper = list()
for word in medicineKeyWords:
    medicineKeyWordsUpper.append(word[0].upper() + word[1 : ])
    
# list of keywords for the statistics category
statsKeyWords = ['statistics', 'statistics:', 'econometrics', 'econometrics:', 'data analysis', 'Data Analysis', 'probability', 'Data Analytics', 'exploratory', 'Data,']
statsKeyWordsUpper = list()
for word in statsKeyWords:
    statsKeyWordsUpper.append(word[0].upper() + word[1 : ])
    
# list of keywords for the human sciences category
humanKeyWords = ['politics', 'history', 'painting', 'music', 'age', 'philosophy', 'law', 'photography', 'guitar', 'greek', 'roman', 'art', 'Canada']
humanKeyWordsUpper = list()
for word in humanKeyWords:
    humanKeyWordsUpper.append(word[0].upper() + word[1 : ])
    
```

```python
# courses are assigned a category in an ordered fashion.
# To avoid that a course name with words corresponding to different 
# categories or with more than one word corresponding to a category is repeated,
# I put it in a set. Then all the courses in the set are attached a category.
# If a course is already in the set, it is not put back in (by default set behaviour)
courseCategoryDict = {'course' : [], 'category' : []}

# IT
courseCategory = set()
for course in courseList:
    for word in course.split(' '):
        if course not in courseCategory and word in itKeyWords or word in itKeyWordsUpper:
            courseCategory.add(course)
for course in courseCategory:
    courseCategoryDict['course'].append(course)
    courseCategoryDict['category'].append('IT')
            #newCourseList.remove(course)

# maths
courseCategory = set()
for course in courseList:
    for word in course.split(' '):
        if course not in courseCategory and word in mathKeyWords or word in mathKeyWordsUpper:
            courseCategory.add(course)       
for course in courseCategory:
    courseCategoryDict['course'].append(course)
    courseCategoryDict['category'].append('maths')
            #newCourseList.remove(course)

# languages
courseCategory = set()
for course in courseList:
    for word in course.split(' '):
        if course not in courseCategory and word in langKeyWords or word in langKeyWordsUpper:
            courseCategory.add(course)
for course in courseCategory:
    courseCategoryDict['course'].append(course)
    courseCategoryDict['category'].append('languages')
            #newCourseList.remove(course)

# psychology
courseCategory = set()
for course in courseList:
    for word in course.split(' '):
        if course not in courseCategory and word in psychKeyWords or word in psychKeyWordsUpper:
            courseCategory.add(course)
for course in courseCategory:
    courseCategoryDict['course'].append(course)
    courseCategoryDict['category'].append('psychology')
            #newCourseList.remove(course)

# economics and management
courseCategory = set()
for course in courseList:
    for word in course.split(' '):
        if course not in courseCategory and word in econKeyWords or word in econKeyWordsUpper:
            courseCategory.add(course)
for course in courseCategory:
    courseCategoryDict['course'].append(course)
    courseCategoryDict['category'].append('economics and management')
            #newCourseList.remove(course)

# sciences
courseCategory = set()
for course in courseList:
    for word in course.split(' '):
        if course not in courseCategory and word in scienceKeyWords or word in scienceKeyWordsUpper:
            courseCategory.add(course)
for course in courseCategory:
    courseCategoryDict['course'].append(course)
    courseCategoryDict['category'].append('sciences')
            #newCourseList.remove(course)

# medicine
courseCategory = set()
for course in courseList:
    for word in course.split(' '):
        if course not in courseCategory and word in medicineKeyWords or word in medicineKeyWordsUpper:
            courseCategory.add(course)
            courseCategoryDict['course'].append(course)
            courseCategoryDict['category'].append('medicine')
            #newCourseList.remove(course)

# statistics
courseCategory = set()
for course in courseList:
    for word in course.split(' '):
        if course not in courseCategory and word in statsKeyWords or word in statsKeyWordsUpper:
            courseCategory.add(course)
for course in courseCategory:
    courseCategoryDict['course'].append(course)
    courseCategoryDict['category'].append('statistics')
            #newCourseList.remove(course)

# humanities
courseCategory = set()
for course in courseList:
    for word in course.split(' '):
        if course not in courseCategory and word in humanKeyWords or word in humanKeyWordsUpper:
            courseCategory.add(course)
for course in courseCategory:
    courseCategoryDict['course'].append(course)
    courseCategoryDict['category'].append('humanities')
```
