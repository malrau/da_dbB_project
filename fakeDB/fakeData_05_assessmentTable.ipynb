<h3>
Create a dictionary of empty lists, where the keys are the attribute names

```python
assessmentSchema = ['assessmentID', 'assessmentType', 'name', 'thresholdScore', 'maxScore']
assessmentDict = dict()
for attribute in assessmentSchema:
    assessmentDict[attribute] = list()
```

<h3>
Load the course table
</h3><br>
Since the assessment is strictly connected to the course, I need to load the course table and exploit this link in order to generate the assessment id and the assessment name. I use the csv module and its <code>reader</code> routine for reading each csv line and assign data to the correct attribute. In particular, when reading lines from the csv file, I specify that the data delimiter is the comma and I assign each read line to a new dictionary where dictionary keys are progressive integers, for easy access. In fact the first line (the one at position <code>0</code>) is the one that stores the attribute names, while the following ones store the actual data.

```python
# select the csv file where the course table data are stored
# and attach each line to a progressive dictionary key for easy access
courseList = list()
courseDict = dict()
key = 0
with open(path + 'tables/coursesTable.csv', newline = '', encoding = 'UTF-16') as csvCourses:
    reader = csv.reader(csvCourses, delimiter = ',')
    for line in reader:
        courseDict[key] = line
        key += 1
```

```python
print(courseDict)
```

Then I rearrange the dictionary so that the keys are the attributes and the values from each previous lines are attached to their correct attribute.

```python
newCourseDict = dict()
for value in courseDict[0]:
    newCourseDict[value] = list()
for i in range(1, len(courseDict)):
    newCourseDict['courseID'].append(int(courseDict[i][0]))
    newCourseDict['discipline'].append(courseDict[i][1])
    newCourseDict['courseName'].append(courseDict[i][2])
    newCourseDict['courseYear'].append(int(courseDict[i][3]))
    newCourseDict['syllabus'].append(courseDict[i][4])
```

<h3>
Generate fake assessment data 
</h3>

```python
# generate random assessment type (chosing among two categories) and name
assessmentID = 0
assessType = ['quiz', 'project']
for courseID, course in zip(newCourseDict['courseID'], newCourseDict['courseName']):
    assessmentID += 1
    assessmentDict['assessmentID'].append(assessmentID)
    assessmentDict['assessmentType'].append(sample(assessType, 1)[0])
    if assessmentDict['assessmentType'][-1] == 'quiz':
        assessmentName = course + ' - Final assessment quiz'
    else:
        assessmentName = course + ' - Final assessment project'
    assessmentDict['name'].append(assessmentName)
    assessmentDict['thresholdScore'].append(60)
    assessmentDict['maxScore'].append(100)
```

```python
with open(path + 'tables/assessmentTable.csv', 'w', newline = '') as csvfile:
    writer = csv.writer(csvfile)
    keys = assessmentDict.keys()
    limit = len(assessmentDict['assessmentID'])
    
    writer.writerow(keys)
    for i in range(0, limit):
        writer.writerow(assessmentDict[k][i] for k in keys)
```
